Cual es el precio medio de alquiler medio de cada uno, la diferencia que hay ¿es estadísticamente significativa?
¿Con que test lo comprobarías?

```r
# Precio medio de alquiler medio de cada Tipo de alquiler

# Aplicamos Shapiro para determinar si la distribución de nuestro dataframe airbnb es normal o gaussiana
# H0 dataframe airbnb proviene de una población normalmente distribuida
# H1 dataframe airbnb no proviene de una población normalmente distribuida    
```

```r
# tapply(airbnb$Precio, airbnb$TipoAlquiler, ks.test(airbnb$Precio, airbnb$TipoAlquiler))
```

```r
# Precio medio de alquiler medio de cada Tipo de alquiler

for (tipoalq in levels(airbnb$TipoAlquiler)){
    print(paste("TipoAlquiler:",tipoalq))
    # Dataframes de airbnb por cada tipo de alquiler
    airbnb_tipoalq <- subset(airbnb,TipoAlquiler == tipoalq)
    print(paste("El precio medio del tipo de alquiler ", 
                tipoalq, "es: ", mean(airbnb_tipoalq$Precio, na.rm = TRUE)))
    
    # stop("sample size must be between 3 and 5000")
    airbnb_tipoalq <- airbnb_tipoalq[sample(1:nrow(airbnb_tipoalq),5000, replace = TRUE),] 
    sh  <- shapiro.test(airbnb_tipoalq$Precio)

    print(paste("Tipo de Alquiler: ",tipoalq,"pvalue, ",  sh$p.value, " W ", sh$statistic))

}

# Conclusión:
# Como todos los p-values son realmente bajos (<0.05) --> rechazamos la H0 (hipótesis nula)
# de manera que aceptamos la hipótesis alternativa H1 dataframe airbnb no proviene de una población normal
```

```r
# Tomo una muestra aleatoria de 5000 elementos para aplicar Shapiro
airbnb_muestra <-airbnb[sample(1:nrow(airbnb),5000),] 
tapply(airbnb_muestra$Precio, airbnb_muestra$TipoAlquiler, shapiro.test)

# Conclusión:

# Los p-values para cada grupo de Tipo de Alquiler están muy por debajo de 0.05 --> 
# rechazamos la hipótesis nula H0, por lo que nuestra distribución no es normal (gaussiana)
```

```r
# Como el dataframe airbnb no sigue una distribución gaussiana o normal --> no podemos aplicar ANOVA
# Y el test que aplicamos es Kruskal-Wallis (aplicable para cualquier distribución)
# H0: 𝜇1=𝜇2=...=𝜇𝑘  (Todos los grupos tienen la misma media)
# H1: at least one pair 𝜇𝑗≠𝜇𝑖 (Al menos la media de un grupo difiere del resto)

kruskal.test( Precio ~ TipoAlquiler, data=airbnb)
tkruskal <-kruskal.test( Precio ~ TipoAlquiler, data=airbnb)

paste ("La diferencia del precio medio por tipo de alquiler es estadisticamente significativa ", 
       "porque el p-value es realmente bajo con p-value=", tkruskal$p.value)

# Conclusión: 
# Como p-value < 2.2e-16, o lo que es lo mismo p-values < 0.00000000000000022
# Significa que p-value es realmente cercano a 0 (< 0.05)--> rechazamos la hipótesis nula H0
# Aceptamos la hipótesis alternativa (al menos la media de uno de los grupos difiere del resto de grupos)
# Concretamente para el tipo de alquiler  Entire home/apt el precio medio difiere del resto de tipos de alquiler
```

Esta claro que las medias de cada uno de estos 5 barrios parecen ser diferentes, pero ¿son estadísticamente diferentes?
¿Que test habría que usar para comprobarlo?

```r
# Aplicamos Shapiro para determinar si la distribución de nuestro dataframe join_airbnb_m2 es normal o gaussiana
# H0 dataframe airbnb proviene de una población normalmente distribuida
# H1 dataframe airbnb no proviene de una población normalmente distribuida    

# Tamaño medio de cada uno de estos 5 Barrios

for (barrio in join_airbnb_m2$Barrio){
    print(paste("Barrio:",barrio))
    # Dataframes de airbnb_entire 
    airbnb_m2 <- subset(airbnb_entire, Barrio == barrio)
    print(paste("El tamaño medio del barrio ", 
                barrio, "es: ", mean(airbnb_m2$MetrosCuadrados, na.rm = TRUE)))
    
    # stop("sample size must be between 3 and 5000")
    airbnb_m2 <- airbnb_m2[sample(1:nrow(airbnb_m2),5000, replace = TRUE),] 
    sh  <- shapiro.test(airbnb_m2$MetrosCuadrados)

    print(paste("Barrio: ",barrio," pvalue, ",  sh$p.value, " W ", sh$statistic))

}

# Conclusión:
# Como todos los p-values son realmente elevados  --> aceptamos la H0 (hipótesis nula) que dice que nuestra
# distribución proviene de una población normal o gaussiana
```

```r
# Podemos aplicar ANOVA
# Y el test que aplicamos es Kruskal-Wallis 
# H0: 𝜇1=𝜇2=...=𝜇𝑘  (Todos los grupos tienen la misma media)
# H1: at least one pair 𝜇𝑗≠𝜇𝑖 (Al menos la media de un grupo difiere del resto)


summary(aov( MetrosCuadrados ~ Barrio, data=airbnb_5b_m2))

# Conclusión el p-value del Barrio es 0.637, realmente elevado --> aceptamos H0
```

```r
oneway.test( MetrosCuadrados ~ Barrio, data=airbnb_5b_m2,var.equal = T)
# Conclusión el p-value del Barrio es 0.6374, realmente elevado --> aceptamos H0
```

```r
# También podemos aplicar el test que aplicamos es Kruskal-Wallis porque es válido para cualquier distribución
# H0: 𝜇1=𝜇2=...=𝜇𝑘  (Todos los grupos tienen la misma media)
# H1: at least one pair 𝜇𝑗≠𝜇𝑖 (Al menos la media de un grupo difiere del resto)

kruskal.test( MetrosCuadrados ~ Barrio, data=airbnb_5b_m2)
tkruskal <-kruskal.test(MetrosCuadrados ~ Barrio, data=airbnb_5b_m2)

paste ("El tamaño medio por barrio no es estadisticamente significativa ", 
       "porque el p-value es realmente alto con p-value=", tkruskal$p.value)

# Conclusión: 
# Como p-value es realmente alto 0.7574 > 0.05 
# Significa que --> aceptamos la hipótesis nula H0 (todos los barrios tienen la misma media de tamaño)
```
